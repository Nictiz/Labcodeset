## Join the LCS and expression map
Inspect some rows.

```python
dfjoin = pd.merge(lcs, dfmap, left_on='LOINC_NUM', right_on='mapTarget')
dfjoin
```

## Filter column names and sort. 

```python
snomedpartmap = dfjoin[['status', 
       'LOINC_NUM', 'COMPONENT', 'PROPERTY', 'TIME_ASPCT', 'SYSTEM',
       'SCALE_TYP', 'METHOD_TYP', 'CLASS', 'NL_LONG_COMMON_NAME',
       'mapTarget', 'expression']]
snomedpartmap
```

## Write to Excel

```python
with pd.ExcelWriter('expressionmap.xlsx') as writer:
    snomedpartmap.to_excel(writer)
```
